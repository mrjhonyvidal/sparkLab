```
from pyspark import SparkContext
```

```
sc = SparkContext()
```

```
rdd = sc.textFile("La+Celestina.txt")
```

```
rdd.takeSample(False, 10)
```

```
palabras = rdd.flatMap(lambda l: l.split(" ")).map(lambda s: s.strip(".,:;?/_! \(\)\'\"")).countByValue()
```
